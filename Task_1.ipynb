```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
df = sns.load_dataset("iris")
```

```python
print("Shape:", df.shape)
```

```python
print("Columns:", df.columns)
```

```python
print(df.head())
```

```python
ax = sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=df)
ax.set_title("Sepal Length vs Petal Length")
```

```python
ax = df["sepal_length"].hist()
ax.set_title("Distribution of Sepal Length")
ax.set_xlabel("Sepal Length")
ax.set_ylabel("Frequency")
```

```python
ax = sns.boxplot(x="species", y="sepal_length", data=df)
ax.set_title("Boxplot of Sepal Length by Species")
```

Insights:

Setosa species ka petal length sab se chhota hai.
Sepal length ka distribution mostly normal hai.
Kuch species me slight outliers present hain.
